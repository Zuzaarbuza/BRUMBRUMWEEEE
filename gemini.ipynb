```python
from google import genai
from google.genai import types

import requests

image_path = "https://goo.gle/instrument-img"
image = requests.get(image_path)

client = genai.Client(api_key="GEMINI_API_KEY")
response = client.models.generate_content(
    model="gemini-2.0-flash-exp",
    contents=["What is this image?",
              types.Part.from_bytes(image.content, "image/jpeg")])

print(response.text)
```

```python
import base64
import mimetypes
from pathlib import Path

pdf_path = r"C:\Users\zdrop\PycharmProjects\BRUMBRUMWEEEE\EASA_AD_2019-0168_1.pdf"

with open(pdf_path, "rb") as pdf_file:
            pdf_data = pdf_file.read()
            encoded_string = base64.b64encode(pdf_data).decode("utf-8")
            
print(encoded_string)
```

```python
from google.genai import types
document = types.Part.from_bytes(data=encoded_string, mime_type="application/pdf")
```

```python
response = client.models.generate_content(
    model="gemini-2.0-flash-exp",
    contents=["What is this pdf?",
              document])

print(response.text)
```
