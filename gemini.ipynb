```python
import os
import time
from google import genai
from dotenv import load_dotenv

load_dotenv()
```

```python
api_key = os.getenv("GEMINI_API_KEY")


client = genai.Client(api_key=api_key)
```

```python
import base64
import mimetypes
from pathlib import Path

pdf_path = r"C:\Users\zdrop\PycharmProjects\BRUMBRUMWEEEE\ADs\AD_2006-0129_1.pdf"

with open(pdf_path, "rb") as pdf_file:
            pdf_data = pdf_file.read()
            encoded_string = base64.b64encode(pdf_data).decode("utf-8")
            
print(encoded_string)
```

```python
from google.genai import types
document = types.Part.from_bytes(data=encoded_string, mime_type="application/pdf")
```

```python
response = client.models.generate_content(
    model="gemini-2.0-flash-exp",
    contents=["What is this pdf?",
              document])

print(response.text)
```
